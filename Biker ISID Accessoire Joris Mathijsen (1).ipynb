Hier maken wij een class aan voor accessoires. Deze koppelen we aan gegevens die wij in het ERD schema hebben vastgesteld vanuit de casus.

```python
#class aanmaken
class Accessoire:

#constructors definiëren

    def __init__ (self):
        self._accessoireId = ''
        self._naam = ''
        self._dagPrijs = ''
        self._borgBedrag = ''
        self._nieuwPrijs = ''
        self._merk = ''
        self._aType = ''
        
# accessoireId attribuut
    def setAccessoireId (self, accessoireId):
        self._accessoireId = accessoireId

    def getAccessoireId (self):
        return self._accessoireId

    accessoireId = property(getAccessoireId, setAccessoireId)
    
    # naam attribuut
    def setNaam (self, naam):
        self._naam = naam

    def getNaam (self):
        return self._naam

    naam = property(getNaam, setNaam)
    
     # dagPrijs attribuut
    def setDagPrijs (self, dagPrijs):
        self._dagPrijs = dagPrijs

    def getDagPrijs (self):
        return self._dagPrijs

    dagPrijs = property(getDagPrijs, setDagPrijs)
    
    # borgBedrag attribuut
    def setBorgBedrag (self, borgBedrag):
        self._borgBedrag = borgBedrag

    def getBorgBedrag (self):
        return self._borgBedrag

    borgBedrag = property(getBorgBedrag, setBorgBedrag)
    
     # nieuwPrijs attribuut
    def setNieuwPrijs (self, nieuwPrijs):
        self._nieuwPrijs = nieuwPrijs

    def getNieuwPrijs (self):
        return self._nieuwPrijs

    nieuwPrijs = property(getNieuwPrijs, setNieuwPrijs)
    
      # merk attribuut
    def setMerk (self, merk):
        self._merk = merk

    def getMerk (self):
        return self._merk

    merk = property(getMerk, setMerk)
    
        # type attribuut
    def setType (self, aType):
        self._merk = aType

    def getMerk (self):
        return self._merk

    merk = property(getMerk, setMerk)
```

Hier maken wij een aantal functies aan die uiteindelijk de gegevens zullen vormen van het accessoire dat we opslaan in de database.
Eerst maken we een willekeurige id aan, en checken wij vanuit de database of die nog niet bestaat. 
Vervolgens pakken wij alle gegevens die de gebruiker invult en slaan we die op in variabelen. Hier wordt nog geverifieerd of de gebruiker wel iets ingevuld heeft.
De laatste functie staat de gebruiker toe om op een knop te drukken en de gegevens in de database op te slaan, waarna de inputs geleegd worden in het GUI.

```python
# functies voor invullen accessoire variabelen

def kiesAccessoireId():
    uniek = 0
    while (uniek == 0):
        accessoireId = randint(111111111, 999999999)
        checkId = cursor.execute("SELECT AccessoireId FROM accessoires").fetchall()
        totalRows = len(checkId)
        for i in range(totalRows):
            if(checkId[i] == accessoireId):
                uniek = 0
            else:
                uniek = 1    
    return accessoireId

def kiesNaam():
    naam = naamEntry.get()
    return naam
       
def kiesDagPrijs():
    dagPrijs = dagPrijsEntry.get()
    return dagPrijs
        
def kiesBorgBedrag():
    borgBedrag = borgBedragEntry.get()
    return borgBedrag
        
def kiesNieuwPrijs():
    nieuwPrijs = nieuwPrijsEntry.get()
    return nieuwPrijs
        
def kiesMerk():
    merk = merkCombo.get()
    alleMerken = cursor.execute("SELECT Naam FROM Merk").fetchall()
    alleMerkCodes = cursor.execute("SELECT MerkCode FROM Merk").fetchall()
    totalRows = len(alleMerken)
    for i in range(totalRows):
        if(alleMerken[i][0] == merk):
            merk = alleMerkCodes[i][0]
            return merk
    return merk
     
def kiesType():
    aType = typeCombo.get()
    alleTypes = cursor.execute("SELECT Naam FROM typeAccessoires").fetchall()
    alleTypeCodes = cursor.execute("SELECT ACode FROM typeAccessoires").fetchall()
    totalRows = len(alleTypes)
    for i in range(totalRows):
        if(alleTypes[i][0] == aType):
            aType = alleTypeCodes[i][0]
            return aType
    return aType
     

#functie aanmaken die input verifieert, fout melding geeft indien niet correct, 
# gegevens opslaat in de database en de input velden leeg maakt.
def maakAccessoire():
        accessoireId = kiesAccessoireId()
        naam = kiesNaam()
        dagPrijs = kiesDagPrijs()
        borgBedrag = kiesBorgBedrag()
        nieuwPrijs = kiesNieuwPrijs()
        merk = kiesMerk()
        aType = kiesType()
        if not naam:
            verificatieLabel.config(text="Er moet een naam ingevuld worden!")
        elif not dagPrijs:
            verificatieLabel.config(text="Er moet een dag prijs ingevuld worden!")
        elif not borgBedrag:
            verificatieLabel.config(text="Er moet een borg bedrag ingevuld worden!")
        elif not nieuwPrijs:
            verificatieLabel.config(text="Er moet een nieuw prijs ingevuld worden!")
        elif not merk:
            verificatieLabel.config(text="Er moet een merk ingevuld worden!")
        elif not aType:
            verificatieLabel.config(text="Er moet een type ingevuld worden!")
        else:
            params = (accessoireId, naam, dagPrijs, borgBedrag, nieuwPrijs,
            merk, aType)
            cursor.execute("INSERT INTO accessoires VALUES (?, ?, ?, ?, ?, ?, ?)", params)
            naamEntry.delete(0, 'end')
            dagPrijsEntry.delete(0, 'end')
            borgBedragEntry.delete(0, 'end')
            nieuwPrijsEntry.delete(0, 'end')
            nieuwPrijsEntry.delete(0, 'end')
            typeCombo.set('')
            merkCombo.set('')
            verificatieLabel.config(text="Accessoire opgeslagen!")
```

Hier maken we een functie aan voor het maken van een tabel. 
Eerst pakken we alle accessoires uit de database en maken we een lege lijst van alle accessoire objecten. 
Daarna maken we een aantal labels die de headers van de tabel gaan vormen, zodat die pas gemaakt worden op het moment dat de gebruiker een tabel wil maken. 
Vervolgens voegen we alle gegevens vanuit de database in de class accessoires toe.
Tot slot maken we een grid met alle onderdelen van de accessoire class.


```python
#tabel maken
def tabelMaken():
    #gegevens verzamelen.
    alleAccessoires = cursor.execute("SELECT AccessoireId, naam, Dagprijs, Borgbedrag, Nieuwprijs, ACode, MerkCode FROM accessoires").fetchall()
    accessoireLijst = []
    maxAccessoires = len(alleAccessoires)
    
    # headers maken voor aankomende tabel.
    tabeldLabel = ttk.Label(root, text="Accessoire ID:")
    tabeldLabel.grid(column=0, row=9, padx=5, pady=5)
    tabelNaamLabel = ttk.Label(root, text="Naam:")
    tabelNaamLabel.grid(column=1, row=9, padx=5, pady=5)
    tabelDagPrijsLabel = ttk.Label(root, text="Dag prijs:")
    tabelDagPrijsLabel.grid(column=2, row=9, padx=5, pady=5)
    tabelBorgBedragLabel = ttk.Label(root, text="Borg bedrag:")
    tabelBorgBedragLabel.grid(column=3, row=9, padx=5, pady=5)
    tabelNieuwPrijsLabel = ttk.Label(root, text="Nieuw prijs:")
    tabelNieuwPrijsLabel.grid(column=4, row=9, padx=5, pady=5)
    tabelMerkLabel = ttk.Label(root, text="Merk:")
    tabelMerkLabel.grid(column=5, row=9, padx=5, pady=5)
    tabelTypeLabel = ttk.Label(root, text="Type:")
    tabelTypeLabel.grid(column=6, row=9, padx=5, pady=5)

    #Gegevens vanuit tabel in class stoppen.
    for n in range(maxAccessoires):
        nieuweAccessoire = Accessoire()
        nieuweAccessoire.accessoireId = alleAccessoires[n][0]
        nieuweAccessoire.naam = alleAccessoires[n][1]
        nieuweAccessoire.dagPrijs = alleAccessoires[n][2]
        nieuweAccessoire.borgBedrag = alleAccessoires[n][3]
        nieuweAccessoire.nieuwPrijs = alleAccessoires[n][4]
        nieuweAccessoire.merk = alleAccessoires[n][5]
        nieuweAccessoire.aType = alleAccessoires[n][6]
        accessoireLijst.append(nieuweAccessoire)
    # Daadwerkelijke tabel aanmaken
    for i in range(maxAccessoires):
        idKolom = Entry(root, width=20, fg='blue',font=('Arial',16,'bold'))
        naamKolom = Entry(root, width=20, fg='blue',font=('Arial',16,'bold'))
        dagPrijsKolom = Entry(root, width=20, fg='blue',font=('Arial',16,'bold'))
        borgBedragKolom = Entry(root, width=20, fg='blue',font=('Arial',16,'bold'))
        nieuwPrijsKolom = Entry(root, width=20, fg='blue',font=('Arial',16,'bold'))
        merkKolom = Entry(root, width=20, fg='blue',font=('Arial',16,'bold'))
        typeKolom = Entry(root, width=20, fg='blue',font=('Arial',16,'bold'))   
        idKolom.grid(row=i +10, column=0)
        idKolom.insert(END, accessoireLijst[i].accessoireId)
        naamKolom.grid(row=i +10, column=1)
        naamKolom.insert(END, accessoireLijst[i].naam)
        dagPrijsKolom.grid(row=i +10, column=2)
        dagPrijsKolom.insert(END, accessoireLijst[i].dagPrijs)
        borgBedragKolom.grid(row=i +10, column=3)
        borgBedragKolom.insert(END, accessoireLijst[i].borgBedrag)
        nieuwPrijsKolom.grid(row=i +10, column=4)
        nieuwPrijsKolom.insert(END, accessoireLijst[i].nieuwPrijs)
        merkKolom.grid(row=i +10, column=5)
        merkKolom.insert(END, accessoireLijst[i].merk)
        typeKolom.grid(row=i +10, column=6)
        typeKolom.insert(END, accessoireLijst[i].aType)
```
