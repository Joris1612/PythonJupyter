Hier volgen de 3 hoofdfuncties van deze applicatie. De gebruiker kan met behulp van 3 knoppen ervoor kiezen om de informatie over de huurinkomsten, inkoop of verkoop te zien. Met behulp van een dropdown menu kan de gebruiker een periode bepalen waarvan de informatie zichtbaar is. Dit geeft de gebruiker de mogelijkheid om van de 3 bovengenoemde elementen alle jaren te bekijken, het laatste jaar (2017) per maand te bekijken, of het laatste jaar per kwartaal in te zien. Er zijn verschillende grafieken en tabellen op basis van de opgevraagde informatie, en er is een fictieve doel lijn op een aantal lijn grafieken toegevoegd.

```python
# functie maken voor huurinkomsten
def huurinkomsten():
    #figuur leegmaken voor het geval dit al gevuld is
    fig.clear()
    
    # periode die de gebruiker wil zien opvragen met behulp van dropdownmenu
    periode = periodeCombo.get()
    
    #gebruiker kiest "alle jaren"
    if periode == 'Alle jaren':
        # subplots aanmaken
        huurinkomstenPJplot = fig.add_subplot(131)
        huurinkomsten_consolePJplot = fig.add_subplot(132)
        huurinkomsten_spelPJplot = fig.add_subplot(133)

        # x en y as vaststellen
        huurinkomstenPJplot.plot(jaren,huurinkomstenPJ, label="Huurinkomsten", color="Green", linewidth=3)
        huurinkomsten_consolePJplot.plot(jaren,huurinkomsten_consolePJ, label="Huurinkomsten", color="Green", linewidth=3)
        huurinkomsten_spelPJplot.plot(jaren,huurinkomsten_spelPJ, label="Huurinkomsten", color="Green", linewidth=3)
        #fictief inkomsten targets opstellen op basis van gemiddelde
        huurinkomstenPJplot.axhline(y=np.mean(huurinkomstenPJ), label="Doel", color="Black", linestyle='--')
        huurinkomsten_consolePJplot.axhline(y=np.mean(huurinkomsten_consolePJ), label="Doel", color="Black", linestyle='--')
        huurinkomsten_spelPJplot.axhline(y=np.mean(huurinkomsten_spelPJ), label="Doel", color="Black", linestyle='--')
        
        # titels aamnaken
        huurinkomstenPJplot.set_title("Totale huurinkomsten")
        huurinkomsten_consolePJplot.set_title("Huurinkomsten console")
        huurinkomsten_spelPJplot.set_title("Huurinkomsten spel")
        fig.suptitle("Huurinkomsten per jaar")

        
        # list maken van plots om herhalende code te voorkomen
        huurinkomstenPJplots = [huurinkomstenPJplot, 
                                huurinkomsten_consolePJplot, 
                                huurinkomsten_spelPJplot]
        
        # loop om design toe te voegen aan alle subplots
        for i in huurinkomstenPJplots:
            i.set_xlabel("Jaar")
            i.set_ylabel("Inkomsten")
            i.grid('on')
            i.set_xlim(2014, 2017)
            i.locator_params(axis="both", integer=True, tight=True)
            i.legend()
            
    
    # gebruiker kiest "laatste jaar per maand"
    elif periode == 'Laatste jaar per maand':
        # subplots aanmaken
        huurinkomstenPMplot = fig.add_subplot(131)
        huurinkomsten_consolePMplot = fig.add_subplot(132)
        huurinkomsten_spelPMplot = fig.add_subplot(133)

        # x en y as vaststellen
        huurinkomstenPMplot.plot(maanden,huurinkomstenPM, 
                                 label="Huurinkomsten", color="Green", linewidth=3)
        huurinkomsten_consolePMplot.plot(maanden,huurinkomsten_consolePM, 
                                 label="Huurinkomsten", color="Green", linewidth=3)
        huurinkomsten_spelPMplot.plot(maanden,huurinkomsten_spelPM,
                                 label="Huurinkomsten", color="Green", linewidth=3)
        #fictief inkomsten targets opstellen op basis van gemiddelde
        huurinkomstenPMplot.axhline(y=np.mean(huurinkomstenPM),
                                    label="Doel", color="Black", linestyle='--')
        huurinkomsten_consolePMplot.axhline(y=np.mean(huurinkomsten_consolePM),
                                    label="Doel", color="Black", linestyle='--')
        huurinkomsten_spelPMplot.axhline(y=np.mean(huurinkomsten_spelPM),
                                    label="Doel", color="Black", linestyle='--')

        # titels aamnaken
        huurinkomstenPMplot.set_title("Totale huurinkomsten")
        huurinkomsten_consolePMplot.set_title("Huurinkomsten console")
        huurinkomsten_spelPMplot.set_title("Huurinkomsten spel")
        fig.suptitle("Huurinkomsten 2017 per maand")

        # list maken van plots om herhalende code te voorkomen
        huurinkomstenPMplots = [huurinkomstenPMplot, 
                                huurinkomsten_consolePMplot, 
                                huurinkomsten_spelPMplot]
        
        # loop voor design
        for i in huurinkomstenPMplots:
            i.set_xlabel("Maand")
            i.set_ylabel("Inkomsten")
            i.grid('on')
            i.set_xlim(1, 12)
            i.locator_params(axis="both", integer=True, tight=True)
            i.legend()

    # gebruiker kiest "laatste jaar per kwartaal"
    else:
        # subplots aanmaken
        huurinkomstenPKplot = fig.add_subplot(131)
        huurinkomsten_consolePKplot = fig.add_subplot(132)
        huurinkomsten_spelPKplot = fig.add_subplot(133)

        # pie-chart plotten
        huurinkomstenPKplot.pie(huurinkomstenPK, 
                                     labels=huurinkomstenPK.values, colors=kleuren, autopct='%1.0f%%')
        huurinkomsten_consolePKplot.pie(huurinkomsten_consolePK, 
                                     labels=huurinkomsten_consolePK.values, colors=kleuren, autopct='%1.0f%%')
        huurinkomsten_spelPKplot.pie(huurinkomsten_spelPK,
                                     labels=huurinkomsten_spelPK.values, colors=kleuren, autopct='%1.0f%%')
        # titels aamnaken
        huurinkomstenPKplot.set_title("Totale huurinkomsten")
        huurinkomsten_consolePKplot.set_title("Huurinkomsten console")
        huurinkomsten_spelPKplot.set_title("Huurinkomsten spel")

        fig.suptitle("Huurinkomsten 2017 per kwartaal")
        huurinkomstenPKplot.legend(huurinkomstenPKplot, labels = kwartaal, loc=2)

        

    #canvas updaten
    canvas.draw_idle()

```

```python
def verkoop():

    #figuur leegmaken voor het geval dit al gevuld is
    fig.clear()
    
    # periode die de gebruiker wil zien opvragen met behulp van dropdownmenu
    periode = periodeCombo.get()
    
    #gebruiker kiest "alle jaren"
    if periode == 'Alle jaren':
        # gegevens voor bar diagram
        bars = np.arange(2014, 2018, 1)
        width = 0.1
        # subplots aanmaken
        verkoopPJplot = fig.add_subplot(131)
        verkoopSpelPJplot = fig.add_subplot(132)
        verkoopConsolePJplot = fig.add_subplot(133)

        # x en y as vaststellen
        verkoopPJplot.plot(jaren,verkoopPJ, label="Verkoop", color="Green", linewidth=3)
        verkoopSpelPJplot.plot(jaren,verkoop_spelPJ, label="Verkoop", color="Green", linewidth=3)
        
        #bar plot vaststellen
        bar1 = verkoopConsolePJplot.bar(bars + width, verkoop_microsoftPJ, width, label='Microsoft')
        bar2 = verkoopConsolePJplot.bar(bars - width, verkoop_sonyPJ, width, label='Sony')
        bar3 = verkoopConsolePJplot.bar(bars, verkoop_nintendoPJ, width, label='Nintendo')
        
        #Fictief target voor verkoop
        verkoopPJplot.axhline(y=np.mean(verkoopPJ),
                                    label="Doel", color="Black", linestyle='--')
        verkoopSpelPJplot.axhline(y=np.mean(verkoop_spelPJ),
                                    label="doel", color="Black", linestyle='--')
        # titels aamnaken
        verkoopPJplot.set_title("Totale verkoop")
        verkoopSpelPJplot.set_title("Verkoop spellen")
        verkoopConsolePJplot.set_title("Verkoop consoles")
        fig.suptitle("Verkoop per jaar")

        # list maken van plots om herhalende code te voorkomen
        verkoopPJplots = [verkoopPJplot, 
                                verkoopSpelPJplot]
        
        # loop voor algemene design
        for i in verkoopPJplots:
            i.set_xlabel("Jaar")
            i.set_ylabel("Verkoop")
            i.grid('on')
            i.set_xlim(2014, 2017)
            i.locator_params(axis="both", integer=True, tight=True)
            i.legend()
        
        #Bardiagram design
        verkoopConsolePJplot.set_xlabel("Jaar")
        verkoopConsolePJplot.set_ylabel("Verkoop")
        verkoopConsolePJplot.grid('on')
        #legenda genereren
        verkoopConsolePJplot.legend()
        
    #gebruiker kiest voor "laatste jaar per maand"
    elif periode == 'Laatste jaar per maand':
        #gegevens voor bar diagram
        bars = np.arange(1, 13, 1)
        width = 0.3
        # subplots aanmaken
        verkoopPMplot = fig.add_subplot(131)
        verkoopSpelPMplot = fig.add_subplot(132)
        verkoopConsolePMplot = fig.add_subplot(133)

        # x en y as vaststellen
        verkoopPMplot.plot(maanden,verkoopPM, label="Verkoop", color="Green", linewidth=3)
        verkoopSpelPMplot.plot(maanden,verkoop_spelPM, label="Verkoop", color="Green", linewidth=3)
        bar1 = verkoopConsolePMplot.bar(bars, verkoop_microsoftPM, width, label='Microsoft')
        bar2 = verkoopConsolePMplot.bar(bars, verkoop_sonyPM, width, label='Sony')
        bar3 = verkoopConsolePMplot.bar(bars, verkoop_nintendoPM, width, label='Nintendo')
        
        #Fictief target voor verkoop
        verkoopPMplot.axhline(y=np.mean(verkoopPM),
                                    label="Doel", color="Black", linestyle='--')
        verkoopSpelPMplot.axhline(y=np.mean(verkoop_spelPM),
                                    label="doel", color="Black", linestyle='--')

        # titels aamnaken
        verkoopPMplot.set_title("Totale verkoop")
        verkoopSpelPMplot.set_title("Verkoop spellen")
        verkoopConsolePMplot.set_title("Verkoop consoles")
        fig.suptitle("Verkoop 2017 per maand")

        # list maken van plots om herhalende code te voorkomen
        verkoopPMplots = [verkoopPMplot, 
                                verkoopSpelPMplot]
        
        # loop voor algemene design
        for i in verkoopPMplots:
            i.set_xlabel("Maand")
            i.set_ylabel("Verkoop")
            i.grid('on')
            i.set_xlim(1, 12)
            i.locator_params(axis="both", integer=True, tight=True)
            i.legend()       
            
        #Bardiagram design
        verkoopConsolePMplot.set_xlabel("Maand")
        verkoopConsolePMplot.set_ylabel("Verkoop")
        verkoopConsolePMplot.grid('on')
        verkoopConsolePMplot.locator_params(axis="both", integer=True, tight=True)
        verkoopConsolePMplot.set_xlim(1, 12)

        #legenda genereren
        verkoopConsolePMplot.legend()
     
    #gebruiker kiest voor "Laatste jaar per kwartaal"
    else:
        #gegevens voor bar diagram
        bars = np.arange(1, 4, 1)
        width = 0.5
        # subplots aanmaken
        verkoopPKplot = fig.add_subplot(131)
        verkoopSpelPKplot = fig.add_subplot(132)
        verkoopConsolePKplot = fig.add_subplot(133)
        

        
        # x en y as vaststellen
        verkoopPKplot.plot(kwartaal,verkoopPK, label="Verkoop", color="Green", linewidth=3)
        verkoopSpelPKplot.plot(kwartaal,verkoop_spelPK, label="Verkoop", color="Green", linewidth=3)
        bar1 = verkoopConsolePKplot.bar(bars + width, verkoop_microsoftPK, width, label='Microsoft')
        bar2 = verkoopConsolePKplot.bar(bars - width, verkoop_sonyPK, width, label='Sony')
        bar3 = verkoopConsolePKplot.bar(bars, verkoop_nintendoPK, width, label='Nintendo')
        

        
        # titels aanmaken
        verkoopPKplot.set_title("Verkoop laatste kwartaal")
        verkoopSpelPKplot.set_title("Verkoop spellen laatste kwartaal")
        verkoopConsolePKplot.set_title("Verkoop consoles laatstek wartaal")

         # pie-chart plotten
        verkoopPKplot.pie(verkoopPK, 
                                     labels=verkoopPK.values, colors=kleuren, autopct='%1.0f%%')
        verkoopSpelPKplot.pie(verkoop_spelPK, 
                                     labels=verkoop_spelPK.values, colors=kleuren, autopct='%1.0f%%')
        # titels aanmaken
        verkoopPKplot.set_title("Totale verkoop")
        verkoopSpelPKplot.set_title("Verkoop spel")
        verkoopConsolePKplot.set_title("Verkoop console")
        fig.suptitle("Verkoop 2017 per kwartaal")
        verkoopPKplot.legend(verkoopPKplot, labels = kwartaal, loc=2)
        

        #Bardiagram design
        verkoopConsolePKplot.set_xlabel("Jaar")
        verkoopConsolePKplot.set_ylabel("Verkoop")
        verkoopConsolePKplot.grid('on')
        #legenda genereren
        verkoopConsolePKplot.legend()
    
    #canvas updaten
    canvas.draw_idle()
```

Hier zijn tot slot een aantal losse dingen toegevoegd aan het programma. Een functie om het figuur te legen, de knoppen die de gebruiker toe staan om het type informatie te selecteren en de combobox die de gebruiker in staat stelt om de periode te selecteren. Ook is het figuur van matplotlib en het canvas van Tkinter aangemaakt zodat die telkens geupdate kan worden in plaats van steeds nieuwe te maken. Dit met het oog op langdurig gebruik, en voorkomen van memory leak. Tot slot wordt dit allemaal in Tkinter geplaatst.

```python
# functie aanmaken voor het leegmaken van het canvas    
def clear():
    fig.clear()
    canvas.draw_idle()

# Knoppen aanmaken voor het aanroepen van bovenstaande functies
huurinkomsten_button = Button(master = root, 
                     command = huurinkomsten,
                     height = 2, 
                     width = 10,
                     text = "Huurinkomsten")

inkoop_button = Button(master = root, 
                     command = inkoop,
                     height = 2, 
                     width = 10,
                     text = "Inkoop")


verkoop_button = Button(master = root, 
                     command = verkoop,
                     height = 2, 
                     width = 10,
                     text = "Verkoop")
 
clear_button = Button(master = root, 
                      command = clear,
                      height = 2,
                      width = 10,
                      text = "Delete")

#Knoppen in Tkinter GUI plaatsen
huurinkomsten_button.pack()
inkoop_button.pack()
verkoop_button.pack()

# Het figuur aanmaken, met bijbehorende canvas
fig = Figure(figsize = (12, 5),
                 dpi = 100)

canvas = FigureCanvasTkAgg(fig,
                               master = root)  
canvas.draw()
  
# Canvas in Tkinter GUI plaatsen
canvas.get_tk_widget().pack()
  
# Toolbar aanmaken voor MatPLotLib
toolbar = NavigationToolbar2Tk(canvas,
                                   root)
toolbar.update()
  
# Toolbar in Tkinter GUI plaatsen
canvas.get_tk_widget().pack()

#variabalen definieren voor comboboxen
periodeWaarde = StringVar
periodeOpties = ("Alle jaren", "Laatste jaar per maand", "Laatste jaar per kwartaal")
    
#Periode optie combobox
periodeCombo = ttk.Combobox(root, textvariable=periodeWaarde)
periodeCombo.state(["readonly"])
periodeCombo['values'] = periodeOpties
periodeLabel = ttk.Label(root, text="Periode:")
periodeCombo.current(0)
periodeLabel.pack()
periodeCombo.pack()





clear_button.pack()
```

Dit is de main functie van het programma, wat we vullen met de Tkinter mainloop()

```python
# Main aanroepen en vullen met Tkinter rootloop
def main():
    root.mainloop()
if __name__ == '__main__' :
    main()   
```
