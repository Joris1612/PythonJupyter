```python
#random importeren voor een "RNG" en tabulate importeren voor een mooie tabel
from random import *
from tabulate import tabulate

#lijst van variabelen maken zodat we die later kunnen gebruiken voor attributen
maatLijst = ["Small", "Medium", "Large"]
kleurenLijst = ["Blauw", "Zwart", "Wit"]

#class aanmaken
class Accessoire:

#constructors definiëren

    def __init__ (self):
        self._type = ''
        self._maat = ''
        self._kleur = ''
        self._serieNummer = ''
        self._verhuurd = ''

 # serienummer attribuut
    def setSerieNummer (self, serieNummer):
        self._serieNummer = serieNummer

    def getSerieNummer (self):
        return self._serieNummer

    serieNummer = property(getSerieNummer, setSerieNummer)
    
# type attribuut
    def setType (self, type):
        self._type = type

    def getType (self):
        return self._type

    type = property(getType, setType)

# maat attribuut
    def setMaat (self, maat):
        self._maat = maat

    def getMaat (self):
        return self._maat

    maat = property(getMaat, setMaat)

# kleur attribuut
    def setKleur (self, kleur):
        self._kleur = kleur

    def getKleur (self):
        return self._kleur

    kleur = property(getKleur, setKleur)

# verhuurd attribuut
    def setVerhuurd (self, verhuurd):
        self._verhuurd = verhuurd

    def getVerhuurd (self):
        return self._verhuurd

    verhuurd = property(getVerhuurd, setVerhuurd)


```

Hier beginnen we met het definieren van een aantal functies zodat we waardes kunnen koppelen aan onze attributen. Voor het gemak van dit voorbeeld worden deze (een soort van) willekeurig gecreëerd. Het serienummer zou in de database de Primary Key vormen. Het type is een van de twee opties aan accessoireres, de helm of het kinderzitje. De maat wordt gekozen uit een lijst van standaard maten (small, medium en large). Ook de kleur wordt gekozen uit een lijst van standaar kleuren (blauw, zwart en wit). Tot slot is er een attribuut "verhuurd", wat de mogelijkheid geeft om op te slaan of het artikel op dit moment beschikbaar is voor verhuur.

```python
# genereer een willekeurig nummer met 9 cijfers voor identificatie van het artikel
def kiesSerieNummer():
    serieNummer = randint(111111111, 999999999)
    return serieNummer
#Kies een type op basis van een willekeurig nummer, 1 of 2, met behulp van de random module
def kiesType():
    x = randint(1, 2)
    if(x == 1):
        type = "Helm"
    else:
        type = "Kinderzitje"
    return type
#Kies een willekeurige maat uit de maatLijst met behulp van de choice methode uit de random module
def kiesMaat():
    maat = choice(maatLijst)
    return maat
#Kies een willekeurige kleur uit de kleurenLijst met behulp van de choice methode uit de random module
def kiesKleur():
    kleur = choice(kleurenLijst)
    return kleur
#Bepaal of het artikel verhuurd is op basis van een willekeurig nummer met behulp van de random module
def kiesVerhuurd():
    y = randint(1, 2)
    if(y == 1):
        verhuurd = "Ja"
    else:
        verhuurd = "Nee"
    return verhuurd
#Creëer de functie die een lijst aanmaakt van 5 accessoires, en de 5 variabelen in de accessoires vult met behulp-
#van de bovenstaande functies
def maakAccessoireLijst():
    accessoireLijst = []
    for accessoire in range(5):
        nieuweAccessoire = Accessoire()
        nieuweAccessoire.serieNummer = kiesSerieNummer()
        nieuweAccessoire.type = kiesType()
        nieuweAccessoire.maat = kiesMaat()
        nieuweAccessoire.kleur = kiesKleur()
        nieuweAccessoire.verhuurd = kiesVerhuurd()
        accessoireLijst.append(nieuweAccessoire)
    return accessoireLijst
    
def createTable():
    table = []
    for x in range(5):
        nieuweRij = [[accessoireLijst[x].serieNummer], [accessoireLijst[x].type], [accessoireLijst[x].maat], [accessoireLijst[x].kleur], [accessoireLijst[x].verhuurd]]
        table.append(nieuweRij)
    return table
        
        
```

Tot slot komt de hoofdfunctie van dit project. In main maken wij een lijst van 5 accessoires door de functie aan te roepen die we hierboven gemaakt hebben. Vervolgens wordt een functie aangemaakt om een tabel te creëeren, waarbij we alle accesoirelijsten in tabel rijen voegen. Dit wordt vervolgens uitgeprint.

```python
def main():
    accessoireLijst = maakAccessoireLijst()
    def accessoireTabelMaken():
        table = [['Serienummer', 'Type', 'Maat', 'Kleur', 'Vehuurd']]
        for x in range(5):
            nieuweRij = [[accessoireLijst[x].serieNummer], [accessoireLijst[x].type], [accessoireLijst[x].maat], [accessoireLijst[x].kleur], [accessoireLijst[x].verhuurd]]
            table.append(nieuweRij)
        return table
    tabel = accessoireTabelMaken()
    print(tabulate(tabel, headers='firstrow', tablefmt='fancy_grid'))
if __name__ == '__main__' :
    main()        
    
```
